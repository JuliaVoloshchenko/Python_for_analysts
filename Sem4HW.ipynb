## 3 задача

Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'view', 
                  'condition', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']].corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
#Если рассматривать цену дома, то не найдена корреляция с этажностью, видом, площадью подвала, а также состоянием, 
#годом постройки и реновации. 
#Зато цена сильно зависит от жилой площади и площади чердака (т.к. много одноэтажных домов), в меньшей степени от 
#количества спален и ванных.
```

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg')
plt.xlabel('Жилая площадь')
plt.ylabel('Цена дома');
```

```python
 #Имеется выраженная зависимость цены от жилой площади, в нижней части можно даже сказать прямая линейная зависимость
```

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg')
plt.xlabel('Площадь чердака')
plt.ylabel('Цена дома');
```

```python
#Чем больше площадь чердака, тем дороже дом
```

```python
sns.boxplot(x=df['bedrooms'], y=df['price'], whis=1.5)
plt.xlabel('Спальни')
plt.ylabel('Цена')
plt.title('Зависимость цены от количества спален');
```

```python
#Самые дорогие дома с 8 спальнями
```

```python
sns.boxplot(x=df['bathrooms'], y=df['price'], whis=1.5)
plt.xlabel('Ванные комнаты')
plt.ylabel('Цена')
plt.title('Зависимость цены от количества ванных комнат')
plt.xticks(rotation=90);
```

```python
#Дороже дома с большим количеством ванных комнат
```
